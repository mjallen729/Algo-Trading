## Merge all the individual tokens into one preprocessing-ready dataset
All of our tokens are tradable on Alpaca and we have historical data spanning back years to present day.  
* 18 tokens total - btc & eth = 16 tradable tokens

```python
import pandas as pd
import glob
import os
```

```python
# Directory containing the CSV files
tokens_dir = "../data/archive/tokens_historical"

# Get all CSV files in the directory
csv_files = glob.glob(os.path.join(tokens_dir, "*.csv"))

print(f"Found {len(csv_files)} CSV files")

# List to store individual dataframes
dfs = []

# Loop through each CSV file
for file in csv_files:
    print(f"Processing: {os.path.basename(file)}")

    # Read CSV, skipping the first row (metadata: url) and using the second row as header
    df_temp = pd.read_csv(file, skiprows=1)
    dfs.append(df_temp)

# Concatenate all dataframes
combined_df = pd.concat(dfs, ignore_index=True)

print(f"\nCombined dataframe shape: {combined_df.shape}")
```

```python
# Drop all Volume columns except Volume USD
volume_cols_to_drop = [col for col in combined_df.columns if 'Volume' in col and col != 'Volume USD']
combined_df = combined_df.drop(columns=volume_cols_to_drop)

print(f"Dropped {len(volume_cols_to_drop)} volume columns")
print(f"Remaining columns: {combined_df.columns.tolist()}")
combined_df.info()
```

```python
# Prep the data fields
new_df.rename({'token': 'symbol', 'volume': 'Volume USD'}, axis=1, inplace=True)

new_df['date'] = pd.to_datetime(new_df['date'])
new_df['unix'] = new_df['date'].astype('int64') // 10**9
new_df['symbol'] = new_df['symbol'] + '/USD'

combined_df.drop(columns=['date'], inplace=True)
new_df.drop(columns=['date'], inplace=True)
```

```python
# Add a merge key
combined_df['uid'] = combined_df['symbol'].str.strip() + combined_df['unix'].astype(str)
new_df['uid'] = new_df['symbol'].str.strip() + new_df['unix'].astype(str)
```

```python
# Keep only rows from new_df that aren't in combined_df
new_data_only = new_df[~new_df['uid'].isin(combined_df['uid'])]

# Concatenate them
final_df = pd.concat([combined_df, new_data_only], ignore_index=True)
final_df.head(5)
```

```python
final_df.shape
```

```python
final_df['symbol'].value_counts()
```

```python
# Filter to keep only tokens with 30k+ samples for deep model
token_counts = final_df['symbol'].value_counts()
tokens_to_keep = token_counts[token_counts >= 30000].index.tolist()

tft_df = final_df[final_df['symbol'].isin(tokens_to_keep)].copy()
```

```python
# Old training data has 1.2M samples
tft_df.shape
```

```python
# Add the date column back
tft_df['date'] = pd.to_datetime(final_df['unix'], unit='s')
```

```python
tft_df = tft_df[
  ['uid', 'unix', 'date', 'symbol', 'open', 'high', 'low', 'close', 'Volume USD']
]
tft_df.rename({'Volume USD': 'volume usd'}, inplace=True)
tft_df.sort_values(by='uid', inplace=True, ignore_index= True)
```

```python
tft_df.head()
```

```python
# Calculate time difference between consecutive rows for each token
tft_df['time_diff'] = tft_df.groupby('symbol')['unix'].diff()

# Find gaps (not 3600 seconds = 1 hour)
gaps = tft_df[tft_df['time_diff'] != 3600].copy()
gaps = gaps[gaps['time_diff'].notna()]  # Remove first row of each symbol (NaN diff)

# Calculate gap size in hours
gaps['gap_hours'] = gaps['time_diff'] / 3600

print(f"Total gaps found: {len(gaps)}")
print(f"\nGaps by token:")
print(gaps.groupby('symbol').size().sort_values(ascending=False))
print(f"\n\nLargest gaps (showing where and how large):")
gaps = gaps[['symbol', 'unix', 'date', 'time_diff', 'gap_hours']].sort_values(by='gap_hours', ascending=False, ignore_index=True)
gaps.head(10)
```
