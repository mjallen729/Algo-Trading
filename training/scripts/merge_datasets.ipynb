## Merge all the individual tokens into one preprocessing-ready dataset
All of our tokens are tradable on Alpaca and we have historical data spanning back years to present day.  
* 18 tokens total - btc & eth = 16 tradable tokens

```python
import pandas as pd
import glob
import os
```

```python
# Directory containing the CSV files
tokens_dir = "../data/archive/tokens_historical"

# Get all CSV files in the directory
csv_files = glob.glob(os.path.join(tokens_dir, "*.csv"))

print(f"Found {len(csv_files)} CSV files")

# List to store individual dataframes
dfs = []

# Loop through each CSV file
for file in csv_files:
    print(f"Processing: {os.path.basename(file)}")

    # Read CSV, skipping the first row (metadata: url) and using the second row as header
    df_temp = pd.read_csv(file, skiprows=1)
    dfs.append(df_temp)

# Concatenate all dataframes
combined_df = pd.concat(dfs, ignore_index=True)

print(f"\nCombined dataframe shape: {combined_df.shape}")
```

```python
# Drop all Volume columns except Volume USD
volume_cols_to_drop = [col for col in combined_df.columns if 'Volume' in col and col != 'Volume USD']
combined_df = combined_df.drop(columns=volume_cols_to_drop)

print(f"Dropped {len(volume_cols_to_drop)} volume columns")
print(f"Remaining columns: {combined_df.columns.tolist()}")
combined_df.info()
```

```python
combined_df.sort_values(by=['symbol', 'unix'], ascending= True, inplace= True)
```
