```python
import pandas as pd
```

```python
df = pd.read_csv('../data/crypto_data.csv')

df.head()
```

```python
df.info()
```

Data has 1.12 million samples

```python
df['date'] = pd.to_datetime(df['date'])
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day
```

```python
df.head(10)
```

```python
c = df.columns
df['time_idx'] = df.groupby('symbol').cumcount()
df = df.reindex(columns=['time_idx', *c])
df.drop(columns=['date'], inplace=True)
```

```python
df.to_csv('../data/crypto_processed.csv', index=False)
```
